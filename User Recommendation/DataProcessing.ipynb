## Drop by number of rating

```python
def drop_by_count(df, col, num):
    counts_col = df.groupby(col, sort=False)[col].transform('size')
    mask = (counts_col > num)
    return df[mask]

rating = drop_by_count(drop_by_count(rating, 'movieId', 10), 'userId', 10)

rating
```

```python
rating_table = rating.pivot_table(index='userId', columns='movieId', values='rating')
rating_table
```
