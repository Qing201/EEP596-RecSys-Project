## Categorical Data Processing

```python
df = pd.read_csv('movie.csv')
df.set_index('tconst', inplace = True)
df
```

```python
all_genres=set()
for line in df['genres']:
    line_s=line.split(",")
    for gen in line_s:
        all_genres.add(gen)
print(all_genres)
```

```python
all_genres_df=pd.DataFrame(0,index=df.index,columns=list(all_genres))
for i in range(len(df.index)):
    gens=df['genres'][i].split(",")
    for gen in gens:
        all_genres_df[gen][i]=1
all_genres_df
    
```

```python
movies = pd.concat([df, all_genres_df], axis=1)
movies.drop(columns='genres', inplace=True)
movies
```

```python
df_ids=pd.read_csv('id.csv')
ids=df_ids['0'].to_list()
```

```python
df_links['imdbId'] = df_links['imdbId'].astype(str)

df_links['imdbId'] = df_links['imdbId'].str.zfill(7)

df_links['imdbId']
```

```python
df_imbd=df_links[df_links['movieId'].isin(ids)]['imdbId']
df_imbd
```

```python
s = [i for i in range(movies.shape[0])]
t_cons=movies.index
movies.insert(1,"Index",s,True)
```

```python
t_cons.to_list
movies.insert(1,"T_Const",t_cons,True)
movies
```

```python
movies.set_index("Index", inplace = True)
movies
```

```python
movies['T_Const']=movies['T_Const'].str[2:]
```

```python
movies
```

```python
df_all=movies[movies['T_Const'].isin(df_imbd)]
df_all
```

```python
T_Const=df_all['T_Const'].to_list()
```

```python
df_links['imdbId']
df_movie_ids=df_links[df_links['imdbId'].isin(T_Const)]['movieId']
df_movie_ids=df_movie_ids.to_list()
df_all.insert(1,'movieId',df_movie_ids,True)
df_all.reset_index(drop=True, inplace=True)
df_all
```

### Normalize

```python
def movies_normalize(movies):
    df = movies.copy()
    movies_norm_col =  df[['startYear', 'averageRating', 'numVotes']]
    df[['startYear', 'averageRating', 'numVotes']] = (movies_norm_col - movies_norm_col.mean()) / (movies_norm_col.max() - movies_norm_col.min())# movies_norm_col.std()

    return df


normalized_movies = movies_normalize(df_all)
normalized_movies
```

```python
normalized_movies.to_csv('../Data/UserRec/IMDB_movie.csv', index=False)
```
